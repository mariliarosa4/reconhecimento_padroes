```python
#Importe as bibliotecas necessárias
from numpy import *
import numpy as np
import pandas as pd
#from sklearn import datasets
import matplotlib.pyplot as plt

from collections import Counter
```

```python
# Importe o arquivo survey.csv

survey = pd.read_csv("./survey.csv")
```

```python
# Quais são os atributos do dataset?
survey.columns 
```

```python
#Qual a característica descritiva de todos atributos?
print(survey.describe(include='all'))
# Quais atributos são categóricos?
categoricos = survey.select_dtypes(include=['object']).copy()
categoricosx = categoricos.astype(str)
# Para os categóricos, quantos elemnetos distintos?
```

```python
# Liste os valores mínimos, máximos, médios e mediana do(s) atributo(s) categórico(s)
print(categoricosx.max())
print(categoricosx.min())
print(categoricosx.mean())    
print(categoricosx.median())

```

```python
# Utilizando lista, mostre quantos valores distintos há para cada atributo
quantidade = []

for item in categoricosx:
    lista = categoricosx[item].tolist()
    quantidade.append(Counter(lista))
i=0;       
for item in categoricosx:
    print(item)
    print(quantidade[i])
    i=i+1
```
